```python
'''JOGO DA VELHA'''

# Aqui estou criando o tabuleiro do jogo e o número que o usuario deverá digitar para a posição que escolher

velha="""               Tabuleiro
   |   |      7 | 8 | 9
---+---+---  ---+---+---
   |   |      4 | 5 | 6
---+---+---  ---+---+---
   |   |      1 | 2 | 3
"""

# Lista de posições (linha e coluna) para cada posição válida do jogo, como ela será imutável, entao cada posicao será uma tupla

posições = [None,(5, 1),(5, 5),(5, 9),(3, 1),(3, 5),(3, 9),(1, 1),(1, 5),(1, 9)]

# Posicões possiveis para exitir um ganhador, linha, colunas e diagonais

ganho = [[ 1, 2, 3],[ 4, 5, 6],[ 7, 8, 9],[ 7, 4, 1],[ 8, 5, 2],[ 9, 6, 3],[ 7, 5, 3],[ 1, 5, 9]]

# Construção do tabuleiro através das lista utilizando o splitlines que retorna uma lista com todas as linhas em string

tabuleiro = []
for linha in velha.splitlines():
    tabuleiro.append(list(linha))

# Pimeiro jogador será o X

jogador = "X" 
jogando = True
jogadas = 0

# Imprime o tabuleiro e o tabuleiro preenchido
while True:
    for t in tabuleiro:  
        print("".join(t))
    if not jogando: 
        break
    if jogadas == 9: 
        
# Se 9 jogadas foram feitas, todas as posições já foram preenchidas
        print("Deu velha...")
        break
        
# Caso o jogador coloque um número maior ou menor entre 1 e 9
    jogada = int(input("Digite a posição entre 1 a 9 (jogador %s):" % jogador))
    if jogada<1 or jogada>9:
        print("Posição inválida")
        continue
        
    # Verifica se a posição está livre
    if tabuleiro[posições[jogada][0]][posições[jogada][1]] != " ":
        print("Posição ocupada.");
        continue
        
# Marca a posição que o jogador escolheu
    
    tabuleiro[posições[jogada][0]][posições[jogada][1]] = jogador
    
# Verfica se houve ganhador
    
    for p in ganho:
        for x in p:
            if tabuleiro[posições[x][0]][posições[x][1]] != jogador:
                break
                
# Se o for terminar sem break, todas as posicoes de p pertencem ao mesmo jogador
        else: 
            print("O jogador %s ganhou (%s): "%(jogador, p))
            jogando = False
            break

            # Alterna jogador
    jogador = "X" if jogador == "O" else "O" 

# Contador de jogadas
    jogadas +=1 
```
